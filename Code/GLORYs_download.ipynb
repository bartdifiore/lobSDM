```python
# Downloading Glorys for Northeast US
# Using Copernicus Marine Toolbox python API

# For Constructing Last day of the month
import datetime
import calendar
import copernicusmarine
import os
import numpy as np

# Access the catalog
catalog = copernicusmarine.describe(overwrite_metadata_cache=True)

# Set box path
box_path = "/Users/aallyn/Library/CloudStorage/Box-Box/"

# # Log in using credentials: 
# copernicusmarine.login()
# Set your credentials as environment variables (or read them securely)
os.environ['CMEMS_USERNAME'] = 'aallyn'
os.environ['CMEMS_PASSWORD'] = 'Maine198513'

# Define the login credentials by reading from environment variables
username = os.getenv('CMEMS_USERNAME')
password = os.getenv('CMEMS_PASSWORD')

# Check directory
os.listdir()

# Variables:
# thetao = potential temperature in C
# so = Salinity in psu
# uo = Eastward ocean current velocity
# vo = Northward ocean current velocity

# Load some years
years = np.arange(1993, 1994, 1)
months = np.arange(1,13)
months = [f"{mnth:02}" for mnth in months]

# checking last day code
# mnth_test = "02"
# datetime.datetime(2010, int(f"{mnth_test}"), calendar.monthrange(2010, int(f"{mnth_test}"))[1], 23, 59, 59).strftime('%Y-%m-%dT%H:%M:%S')

# Loop through them all
for yr in years:
    for mnth in months:

        # Make a stupid end datetime that adapts to the days in each month
        last_day = datetime.datetime(yr, int(f"{mnth}"), calendar.monthrange(yr, int(f"{mnth}"))[1], 23, 59, 59)

        # Download all the months
        copernicusmarine.subset(
        dataset_id = "cmems_mod_glo_phy-thetao_anfc_0.083deg_P1M-m",
        variables = ["thetao"],
        minimum_longitude = -56.2,
        maximum_longitude = -75.8,
        minimum_latitude = 35.2,
        maximum_latitude = 48,
        start_datetime = f"{yr}-{mnth}-01T00:00:00",
        #end_datetime = f"{yr}-{mnth}-31T23:59:59",
        end_datetime = last_day.strftime('%Y-%m-%dT%H:%M:%S'),
        minimum_depth = 0,
        maximum_depth = 1500,
        output_filename = f"NW_Atl_MonthlyTemps_{yr}_{mnth}.nc",
        output_directory = f"{box_path}RES_Data/GLORYs/NW_Atl_MonthlyTemps", 
        force_download = True
        )
```
